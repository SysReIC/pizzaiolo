# Named Pizzas

```python
def showNamedPizzas(df, title, figsize =(3.5,4)):
    display(df[Pizzaiolo.TYPE_COLUMN_NAME].value_counts().sort_index(ascending=False).plot.barh(title=title, figsize=figsize))
    plt.show()
    
showNamedPizzas(full_df, 'NamedPizza : %d' % len(full_df))
# showNamedPizzas(train_df, 'Train - NamedPizza')
# showNamedPizzas(valid_df, 'Valid - NamedPizza')
# showNamedPizzas(test_df, 'Test - NamedPizza')
```

# Concepts

```python
def showConcepts(df, title, figsize=(3.5,4.5)):
    concepts = df[df.columns[4:-6]]
    concepts = concepts.sum()
    concepts = concepts.sort_index(ascending=False)
    bases = df[Pizzaiolo.HASBASE_COLUMN_NAME].value_counts()
    countries = df[Pizzaiolo.HASCOUNTRY_COLUMN_NAME].value_counts().sort_index(ascending=False)
    concepts = pd.concat([countries, concepts, bases])
    # display(concepts)
    display(concepts.plot.barh(title=title, figsize=figsize))
    plt.show()

showConcepts(full_df, 'Instances de Concepts')
# showConcepts(train_df, 'Train Dataset')
# showConcepts(valid_df, 'Valid Dataset')
# showConcepts(test_df, 'Test Dataset')
```
